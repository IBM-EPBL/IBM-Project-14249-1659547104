#Encoding

```python
#Categorical Data : Geography,Gender,Surname
```

```python
dataset['Gender']=dataset['Gender'].astype('category')
dataset['Gender']=dataset['Gender'].cat.codes
dataset.head()
```

```python
dataset['Geography']=dataset['Geography'].astype('category')
dataset['Geography']=dataset['Geography'].cat.codes
dataset.head()
```

```python
dataset['Surname']=dataset['Surname'].astype('category')
dataset['Surname']=dataset['Surname'].cat.codes
dataset.head()
```

```python

X = dataset.iloc[:, :13]
X
```

```python
Y = dataset.iloc[:, 13:]
print(Y)
```

#Scaling

```python
from sklearn.preprocessing import StandardScaler
from sklearn import tree
from sklearn.model_selection import train_test_split
```

```python
sc=StandardScaler()
X=sc.fit_transform(X)
X
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, random_state=0, train_size = .75)
print("Training:\nX:\n")
print(X_train)
print("\nY:\n")
print(Y_train)
print("\nTesting:\nX:\n")
print(X_test)
print("\nY:\n")
print(Y_train)
```
